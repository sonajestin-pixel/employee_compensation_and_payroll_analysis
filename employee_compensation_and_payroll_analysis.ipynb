```python
df['salary_type'] = df['salary_type'].fillna(df['salary_type'].mode()[0])
```

Info:- Missing values in the Salart Type column were replaced using the mode since the variation is categorical and the proportion of missing values was minimal.

```python
df['compulsory_union_code'] = df['compulsory_union_code'].fillna('Not Applicable')
```

Info:- Missing values in the Compulsory Union Code column were filled with "Not Applicable" because union codes are identifier-based categories and using mode could misrepresent the data. 

```python
df['salary_type'].value_counts()
```

Info:- Value counts were computed to examine the distribution of categories in the Salary Type column after handling missing values.

```python
df['compulsory_union_code'].value_counts()
```

Info:- Value counts were calculated to analyze the frequency distribution of different union codes and the newly assigned "Not Applicable" category.

```python
null_percentage = (df.isnull().sum() / len(df)) * 100
null_percentage.sort_values(ascending=False)
```

Info:- The percentage of missing values was calculated for each column should be dropped or imputed. Columns with more than 50% missing values are considered for removal.

```python
# Drop high-missing irrelevant column
df.drop(columns=['post_separation_gross_pay_qtd'],inplace=True)

# Fill money-related columns with 0
df['overtime_gross_pay_qtd'] = df['overtime_gross_pay_qtd'].fillna(0)
df['longevity_gross_pay_qtd'] = df['longevity_gross_pay_qtd'].fillna(0)

# Fill salary using median
df['base_salary'] = df['base_salary'].fillna(df['base_salary'].median())
```

Info:- Columns with extremely high missing values(above 50%) were evaluated. The post_separation_gross_pay_qtd column was dropped due to 98% missing data. Monetary columns such as overtime and longevity pay were filled with 0, as missing values indicate no payment. Base salary missing values were replaced with the median to maintain distribution integrity.

```python
df['employment_status'] = df['termination_year'].apply(lambda x: 'Terminated' if pd.notnull(x) else 'Active')
```

Info:- A new column, employment_status, was created based on the termination_year field. Employees with a recorded termination year were labeled as "Terminated", while those with null values were classified as "Active". This approach preserves the original termination data while clearly identifying employee status.

```python
df['employment_status'].value_counts()
```

```python
null_percentage = (df.isnull().sum() / len(df)) * 100
null_percentage.sort_values(ascending=False)
```

Info:- Termination month and termination year contain approximately 95% null values. However, these were retained because null values indicate active employees, which is critical for attrition analysis.

The dataset was checked for duplicate records using the duplicated() function. No duplicate rows were found, ensuring data uniqueness and integrity.

# DATA VISUALIZATION

```python
import matplotlib.pyplot as plt

plt.hist(df['base_salary'],bins=30,color="green")
plt.title("Distribution of Base Salary")
plt.xlabel("Base Salary")
plt.ylabel("Frequency")
plt.show()
```

 A histogram is used to visualize the distribution of numerical data by grouping values into bins and showing how many observations fall into each bin.

Insights:- 
* The distribution of base salary shows that most employees earn within the mid-salary range(approximately ₹50,000-₹100,000).
* The distribution is positively skewed, as a small number of employees earn significantly higher salaries, extending the tail towards ₹2,50,000.
* This indicates a workforce structure with a large mid-level employee base and fewer high-salary executives.

```python
salary_status = df[['base_salary','employment_status']]

import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x='employment_status',y='base_salary',data=salary_status)
plt.ylim(bottom=0)
plt.title("Salary Distribution by Employment Status")
plt.show()
```

A boxplot displays the central tendency, spread and outliers of a dataset using quartiles.

Insights:-
* The boxplot shows that active employees have a slightly higher median salary compared to terminated employees.
* The salary range for active employees is also wider, indicating greater variability in pay.
* Both groups contain several high-salary outliers, but extreme salaries are more prominent among active employees.
* This suggests that higher-paid employees are more likely to remain active in the organization.

Info:- The lower outliers represent valid low QTD salary values and were retained after verification that no negative values exist.

# CONCLUSION

The overall salary distribution indicates a structured workforce with the majority of employees concentrated within the mid-salary range. The distribution is positively skewed with a small propotion of high earning individuals extending the upper tail. This suggests a hierarchical compensation structure with a strong mid-level employee base and fewer executive-level earners.

Comparative analysis between active and terminated employees reveals that active employees generally have slightly higher median salaries and greater salary variability. This may indicate that higher compensated employees are more likely to remain within the organization. Additionally, the organization demonstrates strong workforce stability, as the majority of employees are currently active, reflecting a relatively low turnover rate.

Salary component analysis shows a weak relationship between base salary and overtime pay suggesting that overtime compensation is largely independent of fixed salary levels. However, a moderate positive relationship exists between overtime and longevity pay indicating that longer-tenured employees may have greater opportunities for additional compensation.

Overall, the findings highlight a stable, predominantely salaried workforce with structured compensation differences across employment categories. The analysis provides valuable insights for HR decision-making related to compensation strategy, workforce planning and employee retention.
